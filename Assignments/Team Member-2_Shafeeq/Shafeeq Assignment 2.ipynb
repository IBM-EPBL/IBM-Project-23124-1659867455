## 6. Find the outliers and replace the outliers

```python
numerical_columns = ["RowNumber", "CustomerId", "CreditScore", "Age", "Tenure", "Balance", 
                     "NumOfProducts", "EstimatedSalary"]
```

```python
outlierss=df.quantile(q=(0.25,0.75))
```

```python
outlierss
```

```python
q1=df[numerical_columns].quantile(0.25)  #(Q1)
q3=df[numerical_columns].quantile(0.75)  #(Q3)
```

```python
IQR=q3-q1
IQR
```

```python
upper_limit= q3 + 1.5*IQR

lower_limit= q1 - 1.5*IQR
```

```python
median = df[numerical_columns].median().astype("int64")
median
```

```python
df[numerical_columns]= np.where(df[numerical_columns]>upper_limit,median,df[numerical_columns])
```

```python
sns.boxplot(df["Age"],color='purple')
```

```python
df["Age"]=np.where(df["Age"]<25,50,df["Age"])
```

```python
sns.boxplot(df["Age"],color='pink')
```
